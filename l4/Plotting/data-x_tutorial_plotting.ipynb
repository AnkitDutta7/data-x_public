# Load Matplotlib

```python
import matplotlib.pyplot as plt # always import pyplot module as plt (standard)
import numpy as np
import pandas as pd

# add inline to print pyplot output withput calling plt.show()
%matplotlib inline 
```

# Simple line plotting

```python
# Simple plot of two hardcoded lists
# plt.plot(x,y)

plt.plot([1,2,3],[-2,5,-8])
plt.show()
```

## Plot 50 random numbers

```python
N = 50 
x = np.arange(1,N+1)
y = np.random.randn(N)

plt.plot(x,y,linestyle='--')
plt.show()
```

# Plotting in the same figure

```python
plt.plot(x,4*y)
plt.plot(x,.5*y)
plt.show()
```

# Scatter plot

```python
# Relationship between two types of variables
plt.scatter(x,y)
plt.show()
```

# Combine line and scatter plots

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,marker='*')
plt.show()
```

# Histograms

```python
vals = np.random.randint(1,10,N)
print sorted(vals)
plt.hist(vals)
plt.show()
```

### Histogram with bins

```python
# histogram with bins
bins = np.linspace(0,10,5)
plt.hist(vals,bins=bins,rwidth=0.8)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2)

# you can also use
# fig = plt.figure()
# ax1 = plt.subplot(221) #2x2 grid, first plot
# ax2 = plt.subplot(222)
# ax3 = plt.subplot(223)
# ax4 = plt.subplot(224)
# ax1.plot(x,y) etc..

ax[0,0].plot(x, y)
ax[1,0].scatter(x, y)
ax[0,1].hist(vals)
ax[1,1].barh(x,np.abs(y))
plt.show()
```

# Add labels, text and change color

```python
# Plot three lines

fig, ax = plt.subplots()

ax.plot(x,y,color='green', linestyle=':', label='line 1',linewidth = 2.5)

ax.plot(x,y*2,'y--', label = 'line 2')

ax.plot(x,y+4,color='black', linestyle='--',marker='o', label = 'line 3')


# labels and legeds are added before calling plt.show()

ax.legend(loc=3) # location of legend is an integer, for text to be correct in the legend the plots need a label

ax.set_title('3 Graphs (g): $\sum^3_i= g_i$\nsubtitle\n',color='blue',fontweight='bold',fontsize=26) # add plot title

ax.set_xlabel('x label') # add xlabel

ax.set_ylabel('y label') # add ylabel


plt.show()
```

# Custom legend

```python
fig, ax = plt.subplots()

l1, = ax.plot(x,y,color='green', linestyle=':', label='line 1',linewidth = 2.5) #Note comma, plot returns tuple object
l2, = ax.plot(x,y*2,'y--', label = 'line 2')
l3, = ax.plot(x,y+4,color='black', linestyle='--',marker='o', label = 'line 3')

plt.figlegend(handles=[l1, l2, l3], labels=['green vert', 'yellow dash', 'black dash dot'],
              loc=[0.6,0.78],frameon=True,numpoints=3)
plt.show()
```

# Double axis plot

```python
fig, ax1 = plt.subplots()

N = 100 # number of samples

err=np.random.randn(N)*5

x=np.linspace(0,10,N)

y=12+x*3.5
y2 = np.sin(2 * np.pi * x)

# plot

ax1.set_xlim(0,10)
ax1.scatter(x,y,color='blue')
ax1.set_ylabel('line',color='blue')
ax1.tick_params('y',colors='blue')


ax2 = ax1.twinx() # create second axis for y2
ax2.plot(x, y2, color='red')
ax2.set_ylabel('sin', color='red')
ax1.set_xlim(0,10)
ax2.tick_params('y', colors='red')
```

# 3d plot

```python
# show matplotlib.org

# Axes 3D needed to set projection='3d'
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Make data, spherical coordinates
u = np.linspace(0, .5*np.pi, 100)
v = np.linspace(0, 2*np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))


# Plot the surface
ax.plot_surface(x, y, z, color='r')

plt.show()
```
